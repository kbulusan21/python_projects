```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
file = open("demo.txt","a")
```

```python
img = Image.open('test.png')
```

```python
if not ".jpg" in img.filename:
    img = img.convert('RGB')
```

```python
div_by = 4
width, height = img.size
#compresses image
new_size = (width//div_by, height//div_by)
compress = img.resize(new_size)

#saves compressed version of image into device
#res_name = "compressed_" + 'test.png'
#compress.save(res_name,optimized=True,quality=50)
```

```python
n_width, n_height = compress.size
print ("Dimensions:", compress.size, "width:", n_width,"height:", n_height)
```

```python
for x in range(0, n_height):
    line = ""
    for y in range(0, n_width):
        pixelRGB = compress.getpixel((y,x))
        R, G, B = pixelRGB
        b = sum([R, G, B]) / 3
        if b < 25.5:
            line +="@"
        elif b < 51:
            line +="%"
        elif b < 76.5:
            line +="#"
        elif b < 102:
            line +="*"
        elif b < 127.5:
            line +="+"
        elif b < 153:
            line +="="
        elif b < 178.5:
            line +="-"
        elif b < 204:
            line +=":"
        elif b < 229.5:
            line +="."
        else:
            line +="." # Use the same default character for both cases
    file.write(line + "\n")
file.close()
```
